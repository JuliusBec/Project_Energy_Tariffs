# Notebook to split Dataset into individual csv file for each user

```python
import pandas as pd
```

```python
import os

# Fix the path - go up two directories to reach project root, then into data folder
file_path = "../../data/Fame4me_11_23-10_24+-+anonym.xlsx"

#no column titles
user_data_df = pd.read_excel(file_path, header=None)
```

```python
#swap rows 0 and 1
user_data_df.iloc[[0, 1]] = user_data_df.iloc[[1, 0]]
print(user_data_df.head())
```

```python
# Skip the first row (column titles) when creating the datetime column
user_data_df.loc[1:, 'datetime'] = pd.to_datetime(user_data_df.loc[1:, 0].astype(str) + ' ' + user_data_df.loc[1:, 1].astype(str), errors='coerce')

#drop datum and bis
user_data_df = user_data_df.drop(columns=[0, 1])
print(user_data_df.head())
```

```python
#create single user df containing datetime, column 0 and 1
single_user_data = user_data_df.iloc[:, [290, 0, 1]]
print(single_user_data.head())

user_name = single_user_data.iloc[1, 1]

print(f"User name: {user_name}")
```

```python
#drop rows 0 and 1 from single_user_data
single_user_data = single_user_data.drop([0, 1])

#rename columns
single_user_data.columns = ['datetime', 'value', 'status']

print(single_user_data.head())
```

```python
# Save the single user data to a new file
single_user_data.to_csv(f"../../data/user_data/user_data_{user_name}.csv", index=False)
```

```python
# repeat the last steps for every individual user
i = 0
while i < 290:
    single_user_data = user_data_df.iloc[:, [290, i, i + 1]]
    user_name = single_user_data.iloc[1, 1]
    #drop rows 0 and 1 from single_user_data
    single_user_data = single_user_data.drop([0, 1])
    #rename columns
    single_user_data.columns = ['datetime', 'value', 'status']
    print(f"User name: {user_name}")
    # Save the single user data to a new file
    single_user_data.to_csv(f"../../data/user_data/user_data_{user_name}.csv", index=False)
    i += 2
```
